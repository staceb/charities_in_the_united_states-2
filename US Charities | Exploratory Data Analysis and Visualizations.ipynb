```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('charity_navigator_clean.csv', index_col = 0)
```

```python
df.drop('charity_url', axis = 1, inplace = True)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
plt.figure(figsize = (15, 8))

sns.heatmap(df.corr(), cmap = 'coolwarm', annot = True)
plt.title('df.corr()');
```

```python
financial_numbers = df.copy()
financial_numbers.drop(['compensation_leader_compensation', 'compensation_leader_expense_percent', 
                       'accountability_score', 'financial_score', 
                       'overall_score'],
                      axis = 1, inplace = True)
```

```python
plt.figure(figsize = (15, 8))

sns.heatmap(financial_numbers.corr(), cmap = 'coolwarm', annot = True)
plt.title('financial_numbers.corr()');
```

```python
sns.pairplot(financial_numbers, palette='coolwarm');
```
